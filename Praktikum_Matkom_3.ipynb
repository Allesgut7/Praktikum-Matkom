```python
from sympy import *
x = symbols('x')

A = 2*x**3 -6*x*2 -7
B = 7*x-8
C = 8*x**4 -6*x+12

derivative_A = A.diff(x)
print("f'(A)=", derivative_A)
derivative_B = B.diff(x)
print("f'(B)=", derivative_B)
derivative_C = C.diff(x)
print("f'(C)=", derivative_C)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')

%matplotlib inline
plt.style.use('seaborn-poster')

def fungsi_A(x) :
  return 2*x**3 -6*x*2 -7

def fungsi_B(x) :
  return 7*x-8

def fungsi_C(x) :
  return 8*x**4 -6*x+12

def Turunan_A(x) :
  return derivative(fungsi_A, x)

def Turunan_B(x) :
  return derivative(fungsi_B, x)

def Turunan_C(x) :
  return derivative(fungsi_C, x)

x = np.linspace(0,np.pi, 10)

fig = plt.figure()

plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)

plt1.plot(x, Turunan_A(x), label ='Turunan A', color = 'blue')
plt1.plot(x, fungsi_A(x), label ='Fungsi A', color = 'red')
plt1.plot(x, Turunan_B(x), label ='Turunan B', color = 'magenta')
plt1.plot(x, fungsi_B(x), label ='Fungsi B', color = 'yellow')
plt1.plot(x, Turunan_C(x), label ='Turunan C', color = 'green')
plt1.plot(x, fungsi_C(x), label ='Fungsi C', color = 'orange')
plt1.grid()
plt1.legend()

plt2.plot(x, Turunan_A(x), label ='Turunan A')
plt2.plot(x, fungsi_A(x), label ='Fungsi A')
plt2.grid()
plt2.legend()

plt3.plot(x, Turunan_B(x), label ='Turunan B')
plt3.plot(x, fungsi_B(x), label ='Fungsi B')
plt3.grid()
plt3.legend()

plt4.plot(x, Turunan_C(x), label ='Turunan C')
plt4.plot(x, fungsi_C(x), label ='Fungsi C')
plt4.grid()
plt4.legend()

fig.subplots_adjust(hspace=.5,wspace=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')

%matplotlib inline
plt.style.use('seaborn-poster')

def integral_a(x) :
  return x**4/2 - 6*x**2 -7*x

def integral_b(x) :
  return 7*x**2/2 - 8*x

def integral_c(x) :
  return 8*x**5/5 - 3*x**2 + 12*x

x = np.linspace(0,np.pi,10)
plt.plot(x, integral_a(x), label = 'Integral A')
plt.plot(x, integral_b(x), label = 'Integral B')
plt.plot(x, integral_c(x), label = 'Integral C')
plt.grid()
plt.legend()
plt.show()
```

```python
#Hitung f(A) dan f(C) [Turunan 2x] !
from sympy import *
x = symbols('x')

A = 2*x**3 -6*x*2 -7
C = 8*x**4 -6*x+12

deri_A = (A.diff(x)).diff(x)
print("f''(A) = ", deri_A)
deri_C = (C.diff(x)).diff(x)
print("f''(C) = ", deri_C)
```

```python
#Hitung F(F(A)) dan F(F(B)) 
from sympy import *
from scipy.integrate import *

x = symbols('x')
a = 2*x**3 -6*x*2 -7
b = 7*x-8
Integrate_a = integrate(integrate(a))
Integrate_b = integrate(integrate(b))

print("F(F(A)) = ", Integrate_a)
print("F(F(B)) = ", Integrate_b)
```

```python
#Plot 2x Integral
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

%matplotlib inline
plt.style.use('seaborn-poster')

x = symbols('x')

def integral_a(x) :
  return x**5/10 - 2*x**3 - 7*x**2/2

def integral_b(x) :
  return 7*x**3/6 - 4*x**2

x = np.linspace(0,np.pi,10)
plt.plot(x, integral_a(x), label = 'F(F(A))')
plt.plot(x, integral_b(x), label = 'F(F(B))')
plt.grid()
plt.legend()
plt.show()
```

```python
#Plot 2x Turunan
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

%matplotlib inline
plt.style.use('seaborn-poster')

x = symbols('x')

def turunan_a(x) :
  return 12*x

def turunan_c(x) :
  return 96*x**2

x = np.linspace(0, np.pi, 5)

plt.plot(x, turunan_a(x), label = 'F"(A)')
plt.plot(x, turunan_c(x), label = 'F"(C)')
plt.grid()
plt.legend()
plt.show()
```
